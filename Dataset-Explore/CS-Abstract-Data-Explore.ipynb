# 1. Read the Data

```python
import pandas as pd
import numpy as np

# Load the data
combined_df = pd.read_csv('dataset/CS-Abstract-Combined-Preprocessed-Dataset.csv')

# Convert the string representation of lists into actual lists of numbers
X = combined_df['extracted_features'].str.replace('\n', '').str.strip('[]').str.split().apply(
    lambda x: [float(i) for i in x])

# Convert lists of numbers into numpy arrays
X = np.array(X.tolist())

# Get the labels as a numpy array
y = np.array(combined_df['label'].tolist())

X, y
```

# 2. Visualize the groups of data

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Create a PCA object
pca = PCA(n_components=2)

# Fit the PCA object to the data
pca.fit(X)

# Transform the data using the fitted PCA object
X_pca = pca.transform(X)

# Plot the data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)
plt.legend()
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA plot of the data')
plt.show()
```

```python
from sklearn.manifold import TSNE

# Create a TSNE object
tsne = TSNE(n_components=2)

# Fit the TSNE object to the data
X_tsne = tsne.fit_transform(X)

# Plot the data
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
plt.legend()
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('TSNE plot of the data')
plt.show()
```
